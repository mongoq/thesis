```python
import cv2
import torch
from PIL import Image
from google.colab.patches import cv2_imshow

# Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')

# Image
! wget -q https://miro.medium.com/max/689/1*RYOxzW7nU4sxmNplxjgxGA.png # motorway image
im1 = cv2.imread('1*RYOxzW7nU4sxmNplxjgxGA.png') 
cv2_imshow(im1)

# Inference
results = model(im1)

# Results
results.print()  
#results.save()
results.show()

#results.xyxy[0]  # im1 predictions (tensor)
#results.pandas().xyxy[0]  # im1 predictions (pandas)
```
